```python
locations = pd.read_csv("CoronaDatasets/locations.csv")
totalcases = pd.read_csv("CoronaDatasets/total_cases.csv")

locations.head()
```

```python
highcases = totalcases.nlargest(15, "World")
```

```python
highcases2 = highcases.nsmallest(15, "World")
highcases2
# I am switching to the smallest so that when I graph the data, it will order it from 4/3- 4/7
```

```python
highcases_highpop = highcases2[["date", "World", "China", "India", "United States", "Indonesia", "Pakistan"]]
highcases_highpop.tail()
```

```python
meltDF = pd.melt(highcases_highpop,
                 id_vars=['date',
                          'World'],
                          
                  value_vars=['China',
                              'India',
                              'United States',
                              'Indonesia',
                              'Pakistan'],
                  var_name ='Country')
```

```python
meltDF
```

```python
highcases_highpop["date"]
```

```python
#Group Bar Chart

c = meltDF["Country"]

x = meltDF["date"]
y = meltDF["value"]
plt.figure(figsize=(15,15))
sns.barplot(x, y, hue = c)
plt.show()
```

```python
#Line plot

c = meltDF["Country"]

x = meltDF["date"]
y = meltDF["value"]
plt.figure(figsize=(15,15))
sns.lineplot(x, y, hue = c)
plt.show()
```
